### Finding heavy traffic patterns on I-94

The goal of our analysis is to determine a few indicators of heavy traffic on I-94. These indicators can be weather type, time of the day, time of the week, etc.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
traffic = pd.read_csv("Metro_Interstate_Traffic_Volume.csv")
```

```python
traffic.head(5)
```

```python
traffic.tail(5)
```

```python
traffic.info()
```

```python
# histogram of traffic_volume column

plt.hist(traffic.loc[:, "traffic_volume"])
plt.show()
```

```python
traffic.loc[:, "traffic_volume"].describe()
```

```python
# correlation between traffic_volume and daytime/nighttime

traffic["date_time"] = pd.to_datetime(traffic["date_time"])
traffic.info()
```

```python
traffic["hour"] = traffic["date_time"].dt.hour
traffic["month"] = traffic["date_time"].dt.month
traffic["day"] = traffic["date_time"].dt.dayofweek
```

```python
traffic.head(5)
```

```python
# Plot traffic_volume by hour

unique_hours = traffic["hour"].unique()
hours_traffic_volume = {}

for hour in unique_hours:
    traffic_by_hour = traffic.loc[:, "hour"] == hour
    hours_traffic_volume[hour] = traffic.loc[traffic_by_hour, "traffic_volume"].mean()
    
```

```python
hours_traffic_volume
```

```python
hours_traffic_volume_series = pd.Series(hours_traffic_volume).sort_index()
hours_traffic_volume_series.plot.bar()
plt.show()
```

```python
# Plot traffic_volume by month

unique_month = traffic["month"].unique()
month_traffic_volume = {}

for month in unique_month:
    traffic_by_month = traffic.loc[:, "month"] == month
    month_traffic_volume[month] = traffic.loc[traffic_by_month, "traffic_volume"].mean()
    
month_traffic_volume_series = pd.Series(month_traffic_volume).sort_index()

month_traffic_volume_series.plot.bar()
plt.show()
```

```python
# Plot traffic_volume by day

unique_day = traffic["day"].unique()
day_traffic_volume = {}

for day in unique_day:
    traffic_by_day = traffic.loc[:, "day"] == day
    day_traffic_volume[day] = traffic.loc[traffic_by_day, "traffic_volume"].mean()
    
day_traffic_volume_series = pd.Series(day_traffic_volume).sort_index()

day_traffic_volume_series.plot.bar()
plt.show()
```

```python
# Isolating the daytime and nighttime data

traffic_daytime = traffic[(traffic["hour"] >= 7) & (traffic["hour"] < 19)]
traffic_nighttime = traffic[(traffic["hour"] >= 19) | (traffic["hour"] < 7)]
```

```python
traffic_daytime["traffic_volume"].describe()
```

```python
traffic_nighttime["traffic_volume"].describe()
```

```python
# plot side by side - daytime and nighttime histograms

plt.figure(figsize=(20, 7))

plt.subplot(1, 2, 1)
plt.hist(traffic_daytime["traffic_volume"])
plt.title("Day time traffic volume")
plt.xlim([0, 8000])
plt.ylim([0, 8500])
plt.xlabel("Traffic Day time")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
plt.hist(traffic_nighttime["traffic_volume"])
plt.title("Night time traffic volume")
plt.xlim([0, 8000])
plt.ylim([0, 8500])
plt.xlabel("Traffic Night time")
plt.ylabel("Frequency")

plt.show()
```

```python
# Day time data is left skewed
# Night time data is right skewed
# Lets just focus on daytime data 
# since nighttime data indicates that traffic is relatively lighter
```

```python
#We're going to look at a few line plots showing 
#how the traffic volume changed according to the following parameters:

#Month
#Day of the week
#Time of day
```

```python
by_month = traffic_daytime.groupby('month').mean()
```

```python
by_month
```

```python
plt.plot(by_month.index, by_month["traffic_volume"])
plt.xticks(rotation=90)
plt.xlabel("month of year")
plt.ylabel("traffic volume mean")
plt.title("Analyze mean traffic volume by month of the year")
plt.show()
```

```python
# traffic is less in nov to feb due to cold weather
# traffic is also less in july surprisingly
```

```python
# line plot by day
by_day = traffic_daytime.groupby('day').mean()
by_day
```

```python
plt.plot(by_day.index, by_day["traffic_volume"])
plt.xlabel("Day of Week")
plt.ylabel("Traffic Volume mean")
plt.title("Analyze traffic volume by day of the week")
plt.show()
```

```python
# plot graph for weekday and weekend by hour
```

```python
weekday = traffic_daytime.loc[traffic_daytime["day"] <= 4]
weekend = traffic_daytime.loc[traffic_daytime["day"] > 4]
```

```python
weekday_by_hour = weekday.groupby("hour").mean()
```

```python
weekend_by_hour = weekend.groupby("hour").mean()
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(1, 2, 1)
plt.plot(weekday_by_hour.index, weekday_by_hour["traffic_volume"])
plt.title("Analyze mean traffic volume by weekday-hourly")
plt.xlabel("Weekday-Hour")
plt.ylabel("Mean traffic volume")
plt.ylim([1000, 7000])

plt.subplot(1, 2, 2)
plt.plot(weekend_by_hour.index, weekend_by_hour["traffic_volume"])
plt.title("Analyze mean traffic volume by weekend-hourly")
plt.xlabel("Weekend-Hour")
plt.ylabel("Mean traffic volume")
plt.ylim([1000, 7000])

plt.show()
```

```python
# we see morning and evening traffic during weekday 
# due to people commuting to/from work
# for weekends we see traffic increasing from noon onwards
```

```python
#The traffic is usually heavier during warm months (March–October) compared to cold months (November–February).
#The traffic is usually heavier on business days compared to weekends.
#On business days, the rush hours are around 7 and 16.
```

```python
# Find correlation for traffic_volume 

traffic_daytime.corr()["traffic_volume"]
```

```python
# we look next at the two categorical weather related columns
# weather_main and weather_description
```

```python
by_weather_main = traffic_daytime.groupby("weather_main").mean()
by_weather_description = traffic_daytime.groupby('weather_description').mean()
```

```python

plt.barh(by_weather_main.index, by_weather_main["traffic_volume"])
plt.title("Mean traffic volume by Weather_Main column")
plt.show()
```

```python
plt.figure(figsize=(7,17))
plt.barh(by_weather_description.index, by_weather_description["traffic_volume"])
plt.title("Mean traffic volume by Weather_Description column")
plt.show()
```

```python
# looks like "Shower Snow" and "Light Rain and Snow" weather descriptions
# have mean traffic volume of > 5000
```

### Conclusion

1. Traffic is more during daytime compared to nighttime
2. Traffic is more during weekdays compared to weekends
3. For daytime and weekdays, traffic is more during commute hours which is 7-8 AM  and 4-5 PM
4. Traffic is also less during winter months - Nov to Feb
5. Traffic is more during Mar to Oct with exception of July

5. 'Shower snow' and 'Light Rain and Snow' weather have impact on traffic as well
6. Even though temp has highest correlation with traffic volume, its not that solid to come to any conclusion
