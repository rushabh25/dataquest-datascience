```python
%%sql
/*
Write a query that returns each genre, with the number of tracks sold in the USA:
in absolute numbers
in percentages.
*/

with usa_invoices as (
    select invoice_line_id from invoice a 
    join invoice_line b on a.invoice_id = b.invoice_id
    where billing_country = 'USA'
)
select a.name as genre_name, 
sum(quantity) as total_sales,
round(sum(quantity) * 100.0 / (select count(*) from usa_invoices),2) as percentages
from genre a join track b on a.genre_id = b.genre_id
join invoice_line c on b.track_id = c.track_id
where invoice_line_id in (select invoice_line_id from usa_invoices)
group by 1
order by 2 desc
```

```python
%%sql

/*
Write a query that finds the total dollar amount of sales 
assigned to each sales support agent within the company. 
Add any extra attributes for that employee that you find are relevant 
to the analysis.
*/
with best_employee as (
select employee_id, sum(total) as total_sales from 
employee a join customer b 
    on a.employee_id = b.support_rep_id
join invoice c on b.customer_id = c.customer_id
group by employee_id
)
select b.total_sales, a.* from employee a join best_employee b on a.employee_id = b.employee_id
order by total_sales desc
```

Looks like hire date had an impact on total_sales.
Those who were hired early had more sales.

```python
%%sql
/*
Write a query that collates data on purchases from different countries.
Where a country has only one customer, collect them into an "Other" group.
The results should be sorted by the total sales from highest to lowest, with the "Other" group at the very bottom.
For each country, include:
- total number of customers
- total value of sales
- average value of sales per customer
- average order value
*/

WITH raw_data AS
(
         SELECT   country,
                  Count(DISTINCT a.customer_id)                             AS total_customers,
                  Sum(total)                                                AS total_sales,
                  Cast(Sum(total) AS FLOAT) / Count(DISTINCT a.customer_id) AS avg_sales_per_customer,
                  Cast(Sum(total) AS FLOAT) / Count(DISTINCT invoice_id)    AS avg_order_value
         FROM     customer a
         JOIN     invoice b
         ON       a.customer_id = b.customer_id
         GROUP BY 1 ) , 
processed_data AS
(
         SELECT
                  CASE
                           WHEN total_customers = 1 THEN 'Other'
                           ELSE country
                  END                         AS country,
                  Sum(total_customers)        AS total_customers,
                  Sum(total_sales)            AS total_sales,
                  Avg(avg_sales_per_customer) AS avg_sales_per_customer,
                  Avg(avg_order_value)        AS avg_order_value
         FROM     raw_data
         GROUP BY 1 ) , 
final_data AS
(
       SELECT *,
              CASE
                     WHEN country = 'Other' THEN 2
                     ELSE 1
              END AS sorter
       FROM   processed_data )
SELECT   country,
         total_customers,
         total_sales,
         avg_sales_per_customer,
            avg_order_value
FROM     final_data
ORDER BY sorter,
         total_sales DESC 
```

```python
%%sql
/*
Write a query that categorizes each invoice as 
either an album purchase or not, and 
calculates the following summary statistics:
Number of invoices
Percentage of invoices
*/

with album_tracks as (
select album_id, count(distinct track_id) as tracks from track
    group by album_id
),
invoice_tracks as (
select invoice_id, b.album_id, count(distinct b.track_id) as tracks
    from invoice_line a join track b on a.track_id = b.track_id
    group by invoice_id, b.album_id
),
non_album_invoices as (
select a.*, b.* from invoice_tracks a join album_tracks b
    on a.album_id = b.album_id where a.tracks != b.tracks
),
album_invoices as (
select a.*, b.* from invoice_tracks a join album_tracks b
    on a.album_id = b.album_id where a.tracks = b.tracks
),
int_table as (
select 'album_purchase' as is_album_purchase, count(distinct invoice_id) as invoices from album_invoices
where invoice_id not in (select invoice_id from non_album_invoices)
union
select 'non-album purchase' as is_album_purchase, count(distinct invoice_id) as invoices from non_album_invoices
)
select *, invoices * 1.0/(select sum(invoices) from int_table) as invoices_percentage from int_table
```
