1. Convert price and odometer to int
2. Convert registration year and month to int
3. Convert postal code to int
4. Convert date_crawled, ad_created and last_seen to datetime

```python
autos.loc[:, "price"] = (autos.loc[:, "price"]
                        .str.replace("$", "")
                        .str.replace(",", "")
                        .astype(int))
```

```python
autos.loc[:, "odometer"] = (autos.loc[:, "odometer"]
                        .str.replace("km", "")
                        .str.replace(",", "")
                        .astype(int))

autos.rename(columns={"odometer": "odometer_km"}, inplace=True)
```

```python
autos_by_brand = autos.loc[:, "brand"].value_counts(normalize=True)
```

```python
# Taking the brands that have atleast 1% of the total values
# so volkswagen to hyundai are brands of our interest

brands_unique_list = autos_by_brand[autos_by_brand > 0.01].index
```

```python
brands_unique_list
```

```python
brands_mean_price = {}

for b in brands_unique_list:
    autos_by_brand = autos.loc[autos.loc[:, "brand"] == b]
    brands_mean_price[b] = autos_by_brand.loc[:, "price"].mean()
```

```python
brands_mean_price
```

```python
# After including the brands that atleast have 1% of the total records
# renault seems to have the least mean price of ~2450
# audi seems to have the max mean price of ~9300
```

```python
brands_mean_mileage = {}

for b in brands_unique_list:
    autos_by_brand = autos.loc[autos.loc[:, "brand"] == b]
    brands_mean_mileage[b] = autos_by_brand.loc[:, "odometer_km"].mean()
```

```python
brands_mean_mileage
```

```python
brands_mean_price_series = pd.Series(brands_mean_price)
brands_mean_mileage_series = pd.Series(brands_mean_mileage)
```

```python
brands_df = pd.DataFrame(brands_mean_price_series, columns=["mean_price"])
```

```python
brands_df["mean_mileage"] = brands_mean_mileage_series
```

```python
brands_df.sort_values(by="mean_mileage", ascending=False)
```
